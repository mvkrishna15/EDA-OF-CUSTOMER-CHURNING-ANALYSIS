```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
               'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
               'TechSupport', 'StreamingTV', 'StreamingMovies']
    
    # Calculate number of rows and columns for subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols
    
    # Create figure and subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
fig.suptitle('Distribution of Services', fontsize=16, y=1.02)
    
    # Flatten axes array for easier iteration
axes_flat = axes.flatten()
    
    # Create countplots
for idx, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes_flat[idx],hue="Churn")
        
        # Customize each subplot
axes_flat[idx].set_title(f'{column} Distribution')
axes_flat[idx].tick_params(axis='x', rotation=45)
        
        # Add count labels on top of each bar
for container in axes_flat[idx].containers:
    axes_flat[idx].bar_label(container)
for idx in range(len(columns), len(axes_flat)):
    fig.delaxes(axes_flat[idx])
plt.tight_layout()
plt.show()
```
